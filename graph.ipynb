```python
df = df.fillna("Unknown")  # Replace NaN values with "Unknown"
df=df.head(50)
df
```

```python
#2
def create_nodes(tx, row):
    query = """
    MERGE (p:Patient {subject_id: $subject_id})
    MERGE (a:Admission {hadm_id: $hadm_id, admission_type: $admission_type})
    MERGE (d:Diagnosis {description: $description})
    MERGE (m:Medication {drug_name: $drug})
    MERGE (t:Test {test_name: $test_name})
    """
    
    tx.run(query, 
           subject_id=row["subject_id"], 
           hadm_id=row["hadm_id"], 
           admission_type=row["admission_type"],
           description=row["description"], 
           drug=row["drug"], 
           test_name=row["test_name"])

# Insert Nodes into Neo4j
with driver.session() as session:
    for _, row in df.iterrows():
        session.execute_write(create_nodes, row)
```

```python
#2
def create_relationships(tx, row):
    query = """
    MATCH (p:Patient {subject_id: $subject_id})
    MATCH (a:Admission {hadm_id: $hadm_id})
    MATCH (d:Diagnosis {description: $description})
    MATCH (m:Medication {drug_name: $drug})
    MATCH (t:Test {test_name: $test_name})

    MERGE (p)-[:HAD_ADMISSION]->(a)
    MERGE (a)-[:DIAGNOSED_WITH]->(d)
    MERGE (a)-[:PRESCRIBED]->(m)
    MERGE (a)-[:UNDERWENT_TEST]->(t)
    """
    
    tx.run(query, 
           subject_id=row["subject_id"], 
           hadm_id=row["hadm_id"], 
           description=row["description"], 
           drug=row["drug"], 
           test_name=row["test_name"])

# Insert Relationships into Neo4j
with driver.session() as session:
    for _, row in df.iterrows():
        session.execute_write(create_relationships, row)

print("✅ Knowledge Graph Successfully Created in Neo4j!")
```

```python
import networkx as nx
import matplotlib.pyplot as plt
from py2neo import Graph

# Connect to Neo4j
neo4j_uri = "neo4j+s://5f06d1b4.databases.neo4j.io"
neo4j_user = "neo4j"
neo4j_password = "Nk30WHZPhuWpLUeMRHyuspzr-xjoX7zWw1kcTm8JKlE"

graph = Graph(neo4j_uri, auth=(neo4j_user, neo4j_password))

# Fetch Graph Data
query = """
MATCH (n)-[r]->(m)
RETURN labels(n) AS from_type, COALESCE(n.subject_id, n.hadm_id, n.description, n.drug_name, n.test_name, "Unknown") AS from,
       type(r) AS relation, 
       labels(m) AS to_type, COALESCE(m.subject_id, m.hadm_id, m.description, m.drug_name, m.test_name, "Unknown") AS to
LIMIT 200
"""
data = graph.run(query).data()

# Print Raw Data for Debugging
print("\n--- Nodes and Relationships Retrieved ---")
for row in data:
    print(f"From: {row['from']} ({row['from_type']}), Relation: {row['relation']}, To: {row['to']} ({row['to_type']})")

# Handle Missing Values
filtered_data = [row for row in data if row["from"] != "Unknown" and row["to"] != "Unknown"]

# Create NetworkX Graph
G = nx.DiGraph()

# ✅ First, add all nodes
nodes = set()
for row in filtered_data:
    from_node = f"{row['from']} ({row['from_type'][0]})"
    to_node = f"{row['to']} ({row['to_type'][0]})"
    nodes.add(from_node)
    nodes.add(to_node)

for node in nodes:
    G.add_node(node)

# ✅ Now, add edges
for row in filtered_data:
    from_node = f"{row['from']} ({row['from_type'][0]})"
    to_node = f"{row['to']} ({row['to_type'][0]})"
    G.add_edge(from_node, to_node, label=row["relation"])

# Generate positions for nodes
pos = nx.spring_layout(G, seed=42, k=0.15)

# ✅ Ensure edge labels only include existing nodes
edge_labels = {}
for row in filtered_data:
    from_node = f"{row['from']} ({row['from_type'][0]})"
    to_node = f"{row['to']} ({row['to_type'][0]})"
    if from_node in pos and to_node in pos:
        edge_labels[(from_node, to_node)] = row["relation"]

# Draw the Graph
plt.figure(figsize=(15, 10))
nx.draw(G, pos, with_labels=True, node_color="lightblue", edge_color="gray", node_size=2500, font_size=8)
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=7)

plt.title("EHR Knowledge Graph from Neo4j")
plt.show()
```

```python
import os
from langchain.chains import GraphCypherQAChain
from langchain.graphs import Neo4jGraph
from langchain.chat_models import ChatOpenAI
from dotenv import load_dotenv

# Load environment variables from .env file (if available)
load_dotenv()

# Directly assign Neo4j credentials if not using .env
NEO4J_URI = os.getenv("NEO4J_URI", "neo4j+s://5f06d1b4.databases.neo4j.io")
NEO4J_USER = os.getenv("NEO4J_USER", "neo4j")
NEO4J_PASSWORD = os.getenv("NEO4J_PASSWORD", "Nk30WHZPhuWpLUeMRHyuspzr-xjoX7zWw1kcTm8JKlE")
OPENAI_API_KEY = os.getenv("OPENAI_API_KEY")

# Check for missing credentials
if not OPENAI_API_KEY:
    raise ValueError(" OpenAI API Key is missing. Set it as an environment variable.")

# Connect to Neo4j
graph = Neo4jGraph(url=NEO4J_URI, username=NEO4J_USER, password=NEO4J_PASSWORD)

# Initialize OpenAI model (GPT-4o)
llm = ChatOpenAI(temperature=0, model="gpt-4o", openai_api_key=OPENAI_API_KEY)

# Allow execution of Cypher queries
qa_chain = GraphCypherQAChain.from_llm(
    llm, 
    graph=graph, 
    verbose=True, 
    allow_dangerous_requests=True
)

def process_query(user_query):
    """ Converts a human text query to Cypher, retrieves data, and returns results. """
    try:
        print("\nProcessing query:", user_query)
        
        # Convert text to Cypher and fetch data
        response = qa_chain.run(user_query)
        
        print("\n🔹 Cypher Query Generated:")
        print(response['cypher_query'])  # Debug: Print generated Cypher query
        
        print("\nQuery Results:")
        print(response['results'])  # Display results
        
        return response['results']
    
    except Exception as e:
        print(f"Error: {str(e)}")
        return None

if __name__ == "__main__":
    while True:
        user_query = input("\nEnter your medical query (or type 'exit' to quit): ")
        if user_query.lower() == "exit":
            print("Exiting chatbot.")
            break
        
        results = process_query(user_query)
        if results:
            print("\n✅ Query Results:", results)
```
